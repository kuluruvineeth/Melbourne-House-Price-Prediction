```python
# importing pandas library
import pandas as pd

# reading data from csv as a pandas dataframe
df = pd.read_csv('Melbourne_housing_corrected.csv')
```

```python
# by default it displays first 5 rows
df.head()
```

```python
# to get particular row lets say 5th row
df.iloc[4]
```

```python
# lets build model using gradient boosting algorithm

from sklearn.model_selection import train_test_split
# ensemble includes gradient boosting algorithm
from sklearn import ensemble
# for performance analysis
from sklearn.metrics import mean_absolute_error
```

```python
# lets scrub our data(i.e modify,cleaning,converting text to numerical values)

# lets delete the columns which we do not want

del df['Address']
del df['Method']
del df['SellerG']
del df['Date']
del df['Postcode']
del df['Latitude']
del df['Longitude']
del df['Regionname']
del df['Propertycount']

```

```python
df.columns
```

```python
df.head()
```

```python
#lets remove missing values
# axis = 0 means drop rows with missing values
# how = any means drop rows or columns with any missing values
# thresh = None means like ex : 4 means remove rows or columns with 4 or more missing values
# subset = None means no specific genre or subset is selected
# inplace = true means if true do update rather than replace


df.dropna(axis = 0,how = 'any',thresh = None,subset = None,inplace = True)
```

```python
# now lets convert columns that contain non-numeric data to numeric values
# using one-hot encoding

df = pd.get_dummies(df,columns=['Suburb','CouncilArea','Type'])
```

```python
df.head()
```

```python
# Now lets assign dependent and independent varibles

# except price column all columns are independent 
X = df.drop('Price',axis=1)
# dependent variable price column alone
y = df['Price']
```

```python
# splitting the data 70/30 

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,shuffle = True)
```

```python
# time to apply our algorithm

### n_estimators states the number of decision trees
### learning rate controls the at which additional decision trees influence the overall prediction
### max_depth defines the maximum number of layers(depth) for each decision true
### min_samples_split defines minimum number of samples required to execute a new binary split
### here min_samples_split = 4 means that there must be four available samples in order to create new branch
### min_samples_leaf says that minimum number of samples that must appear in child node before new branch can be implemented
### max_features is total number of features presented to the model when determining best split
### loss = huber calculates models error rate,it protects against outliers and anomalies
### huber(is combination of least squares regression and least absolute deviations)

### the above all are hyperparameters of gradient boosting
model = ensemble.GradientBoostingRegressor(
    n_estimators = 150,
    learning_rate = 0.1,
    max_depth = 30,
    min_samples_split = 4,
    min_samples_leaf = 6,
    max_features = 0.6,
    loss = 'huber'
)
```

```python
# lets link training data to algorithm model to train prediction model

model.fit(X_train,y_train)
```

```python
mae_train = mean_absolute_error(y_train,model.predict(X_train))
print("Training set mean absolute error: %.2f" %mae_train)
```

```python
mae_test = mean_absolute_error(y_test,model.predict(X_test))
print("Test set mean absolute error: %.2f" %mae_test)
```

```python
# this means that on average the training set miscalculated the actual property value by $28,452.55
# the test set miscalculated the property value by $166,808.57 on average
# conlusion: 
# As many of the properties in the dataset are in excess of seven figues,$28,452.55 constitutes a reasonably low error rate
# the test data error rate is $166,808.57.
# high discrepancy between training and testing data indicates overfitting.
```

```python
# inorder to avoid above overfitting problem lets try to play with parameters

model = ensemble.GradientBoostingRegressor(
    n_estimators = 250,
    learning_rate = 0.1,
    max_depth = 5,
    min_samples_split = 4,
    min_samples_leaf = 6,
    max_features = 0.6,
    loss = 'huber'
)
```

```python
model.fit(X_train,y_train)
```

```python
mae_train = mean_absolute_error(y_train,model.predict(X_train))
print("Training set mean absolute error %0.2f" %mae_train)
```

```python
mae_test = mean_absolute_error(y_test,model.predict(X_test))
print("Test set mean absolute error : %0.2f" %mae_test)
```

```python
# now we can clearly see that the difference between training error and testing error is comparatively less when compared to previous one
```

```python
# (WARNING)this is optional do not try to execute this cell because it needs lot of computational resources
# lets try with one more optimization technique called grid search(which finds out best hyperparameters)

from sklearn.model_selection import GridSearchCV
model = ensemble.GradientBoostingRegressor()

hyperparameters = {
    'n_estimators' : [200,300],
    'max_depth' : [4,6],
    'min_samples_split' : [3,4],
    'min_samples_leaf' : [5,6],
    'learning_rate' : [0.01,0.02],
    'max_features' : [0.8,0.9],
    'loss' : ['ls','lad','huber']
}

grid = GridSearchCV(model,hyperparameters,n_jobs = 1)

grid.fit(X_train,y_train)

grid.best_params_

mae_train = mean_absolute_error(y_train,grid.predict(X_train))
print("Training set mean absolute error %0.2f" %mae_train)

mae_test = mean_absolute_error(y_test,grid.predict(X_test))
print("Test set mean absolute error : %0.2f" %mae_test)
```
